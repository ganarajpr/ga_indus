```python
import pygad
import numpy as np
```

```python
import pandas as pd
text = pd.read_csv('cleaned.csv')

genes = pd.read_csv('indus_alpha.csv')
ph = pd.read_csv('devph.csv')
gene_space = np.arange(len(ph))
gene_space
```

```python
def getPh(index):
    return ph['0'][int(index)]

def getGene(index):
    return genes['letter'][index]
```

```python
sentences = text['split']
sen = []
for i in range(len(sentences)):
    sp = sentences[i].split(',')
    sen.append(sp)
sen[0]
```

```python
from predict import isSanskritWord
```

```python
def fitness_fun(solution, solution_idx):
    ind2san = {}
    for i in range(len(solution)):
       ind2san[getGene(i)] = getPh(i)
    sol = []
    for i in range(len(sen)):
        sansen = []
        for j in range(len(sen[i])):
            sansen.append(ind2san[sen[i][j]])
        word = ''.join(sansen)
        op = 1 if isSanskritWord(word) else 0
        sol.append(op)
    return len(sen) - np.sum(sol)
```

```python
ga_instance = pygad.GA(num_generations=10,
                       num_parents_mating=10,
                       fitness_func=fitness_fun,
                       sol_per_pop=10,
                       num_genes=len(genes),
                       gene_space=gene_space,
                       mutation_percent_genes=0.1,
                       mutation_type="random",
                       mutation_by_replacement=True)
```

```python
ga_instance.run()
```
