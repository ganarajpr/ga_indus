```python
import pandas as pd
indus = pd.read_csv('orderedl2r.csv')
indus
```

```python
def filter_fn(row):
    if '/' in row['split']:
        return False
    else:
        return True
m = indus.apply(filter_fn, axis=1)
df1 = indus[m]
df1

onlySplit = df1['split']
onlySplit.to_csv('cleaned.csv', index=False)
```

```python
chars = list(df1['split'])
```

```python
letters = set([])
for i in range(len(chars)):
    br = chars[i].split(',')
    if i%1000 == 0:
        print("Done:", i)
    letters = letters.union(set(br))
print(letters)
print(len(letters))
```

```python
lt = pd.DataFrame(data=letters,columns=['letter'])
lt
```

```python
lt.to_csv('indus_alpha.csv', index=False)
```
